```python
from sklearn.preprocessing import LabelEncoder
from tqdm import tqdm

def one_hot(df):
    
    df_dict = df.dtypes.to_dict()
    strcols = [col for col in df.columns if df_dict[col] not in ['int64', 'float64']]
    numcols = [col for col in df.columns if df_dict[col] in ['int64', 'float64']]
    print('-'*50)
    print('cols number check:', len(strcols)+len(numcols) == len(df.columns))
    
    encoder_dict = {}
    for col in tqdm(strcols):
        labelencoder = LabelEncoder()
        df[col+'_invsd'] = labelencoder.fit_transform(df[col])
        try:
            df[col+'_invsd'].replace(labelencoder.transform([np.nan])[0], np.nan, inplace=True)
        except Exception as e:
            pass
        encoder_dict[col] = labelencoder
    return df, numcols, [col+'_invsd' for col in strcols]
```

```python
bureau, numcols_['bureau'], strcols_['bureau'] = one_hot(bureau)
```

### 使用featuretools 自动化特征工程开发
它擅长将时间和关系数据集转换为用于机器学习的特征矩阵。

featuretools实际上就是把常见的特征工程方法进行了自动化封装，自动进行常规特征工程操作比如多重的groupby。

特征工程常见的方法分为两种：1、针对单表的transform操作，例如log变换，特征编码等，都是在一张表上进行的；2、groupby聚合操作，一般是跨表进行的，比如groupby min max mean等等。

featuretools涉及到3个概念 实体(entity,多个实体则称为实体集entityset)，关系（relationship），算子(primitive)，所谓的实体就是一张表或者一个dataframe，多张表的集合就叫实体集，关系就是表之间的关联键的定义，而算子就是一些特征工程的函数，例如groupby mean max min等等，featuretools实际上就是提供了一个框架让我们可以方便快速的通过简约的代码来实现单表的转换操作和多表的跨表连接操作。


```python
import featuretools as ft
import warnings
warnings.filterwarnings('ignore')
```

```python
POS_CASH_balance['Bureau_balance_key'] = POS_CASH_balance['SK_ID_PREV'].astype(str)+POS_CASH_balance['MONTHS_BALANCE'].astype(str)
credit_card_balance['credit_card_balance_key'] = credit_card_balance['SK_ID_PREV'].astype(str)+credit_card_balance['MONTHS_BALANCE'].astype(str)
previous_application['previous_application_key'] = previous_application['SK_ID_PREV'].astype(str)+credit_card_balance['MONTHS_BALANCE'].astype(str)
Bureau_balance['Bureau_balance_key'] = Bureau_balance['SK_ID_BUREAU'].astype(str)+Bureau_balance['MONTHS_BALANCE'].astype(str)
installments_payments = installments_payments.reset_index().rename(columns={'index':'install_key'})
```

```python
columns1 = numcols + [col+'_invsd' for col in strcols]
columns1.remove('TARGET')
application = pd.concat([applytrain[columns1], applytest])
```

```python
%%time 
# 查看本机计算资源
import multiprocessing
multiprocessing.cpu_count()
```

```python
for i in tqdm(range(0, 36)):
    print('-'*15, 'begin', '-'*15)
    low, upper = i*10000, (i+1)*10000
    application1 = application.iloc[low:upper]

    bureau1 = bureau[bureau['SK_ID_CURR'].isin(list(application1['SK_ID_CURR']))]
    Bureau_balance1 = Bureau_balance[Bureau_balance['SK_ID_BUREAU'].isin(list(bureau1['SK_ID_BUREAU']))]
    POS_CASH_balance1 = POS_CASH_balance[POS_CASH_balance['SK_ID_CURR'].isin(list(application1['SK_ID_CURR']))]
    credit_card_balance1 = credit_card_balance[credit_card_balance['SK_ID_CURR'].isin(list(application1['SK_ID_CURR']))]
    previous_application1 = previous_application[previous_application['SK_ID_CURR'].isin(list(application1['SK_ID_CURR']))] 
    installments_payments1 = installments_payments[installments_payments['SK_ID_CURR'].isin(list(application1['SK_ID_CURR']))] 
    
    print('-'*15, 'cut finished', '-'*15)
    dataframes = {
        "application": (application1[[col for col in numcols+strcols if col != "TARGET"]], "SK_ID_CURR"),
        "bureau": (bureau1[strcols_['bureau']+numcols_['bureau']], "SK_ID_BUREAU", "SK_ID_CURR"),
        "Bureau_balance": (Bureau_balance1[strcols_['Bureau_balance']+numcols_['Bureau_balance']], "Bureau_balance_key", "MONTHS_BALANCE"),
        "POS_CASH_balance": (POS_CASH_balance1[strcols_['POS_CASH_balance']+numcols_['POS_CASH_balance']], "Bureau_balance_key", "MONTHS_BALANCE"),
        "credit_card_balance": (credit_card_balance1[strcols_['credit_card_balance']+numcols_['credit_card_balance']], "credit_card_balance_key", "MONTHS_BALANCE"),
        "previous_application": (previous_application1[strcols_['previous_application']+numcols_['previous_application']], "SK_ID_PREV"),
        "installments_payments": (installments_payments1, "install_key"),
    }

    relationships = [
        ("application", "SK_ID_CURR", "bureau", "SK_ID_CURR"),
        ("bureau", "SK_ID_BUREAU", "Bureau_balance", "SK_ID_BUREAU"),
        ("application", "SK_ID_CURR", "POS_CASH_balance", "SK_ID_CURR"),
        ("application", "SK_ID_CURR", "credit_card_balance", "SK_ID_CURR"),
        ("application", "SK_ID_CURR", "previous_application", "SK_ID_CURR"),
        ("application", "SK_ID_CURR", "installments_payments", "SK_ID_CURR"),
    ]

    print('-'*15, 'features', '-'*15)
    feature_matrix_sessions, features_defs = ft.dfs(
        dataframes=dataframes
        ,relationships=relationships
        ,target_dataframe_name="application"
        ,n_jobs=-1
        ,agg_primitives = ['sum', 'std', 'max', 'skew', 'min', 'mean', 'count', 'percent_true', 'num_unique', 'mode']
    )

    print('-'*15, 'saving', '-'*15)
    feature_matrix_sessions.to_csv(f"./output/feature_matrix_sessions{i}.csv", sep=',')
```

```python
fea_list = []
for i in tqdm(range(0, 36)):
    fea = pd.read_csv(f"./output/feature_matrix_sessions{i}.csv")
    fea_list.append(fea)
features = pd.concat(fea_list)
```
